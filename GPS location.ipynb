```python
df = pd.read_csv("location_pre_20190113_000000000000.csv", compression='gzip')
```

```python
locations = list(zip(df.latitude.values, df.longitude.values))
```

```python
# From the github Jeff sent
results2 = rg.search(locations)
```

```python
reader = csv.reader(open('Country codes - cc.csv', 'r'))
d = {}
for cc, un_code in reader:
    d[cc] = int(un_code)
```

```python
def cc_to_un(x):
    return d[x]
```

```python
df.cc = df.cc.apply(cc_to_un)
```
